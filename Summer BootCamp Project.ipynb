# First check for outliers

```python
plt.figure(figsize=(10,20))
sns.boxplot(data = data, orient = 'h')
plt.xticks(rotation = 60)
plt.show()
```

```python
def remove_outliers(data, cols):
    for col in cols:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data
```

```python
data_filled = data.fillna(data.median(numeric_only=True))
```

```python
data.head()
```

```python
data_filled['S.F.Ratio'] = pd.to_numeric(data_filled['S.F.Ratio'], errors='coerce')
data_filled['S.F.Ratio'].fillna(data_filled['S.F.Ratio'].median(), inplace=True)
```

```python
numeric_cols = data_filled.select_dtypes(include=['float64', 'int64']).columns
```

```python
data_no_outliers = remove_outliers(data_filled, numeric_cols)
```
